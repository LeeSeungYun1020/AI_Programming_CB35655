```python
%load_ext autoreload
%autoreload 2
from main import KNN
import matplotlib.pyplot as plt
```

```python
for (train_file, test_file) in [["lin_train.txt", "lin_test.txt"], ["nonlin_train.txt", "nonlin_test.txt"]]:
    rmse = []
    ran = range(1, 101)
    for i in ran:
        knn = KNN(i)
        knn.setData("train", train_file)
        knn.setData("test", test_file)
        knn.buildModel()
        knn.testModel()
        rmse.append(knn.calcRMSE())

    plt.title(train_file.split(".")[0])
    plt.plot(ran, rmse)
    min_error = min(rmse)
    best_k = rmse.index(min_error) + 1
    plt.plot(ran, rmse)
    plt.scatter(best_k, min_error, marker='X', s=30, c='red')
    plt.show()
    print("Best k =", best_k)
    print("rmse =", min_error)
```

```python
knn = KNN(6)
knn.setData("train", "lin_train.txt")
knn.setData("test", "lin_test.txt")
knn.buildModel()
knn.testModel()
knn.report()
```
