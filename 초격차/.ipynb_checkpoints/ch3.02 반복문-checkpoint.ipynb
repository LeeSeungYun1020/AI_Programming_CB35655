# 반복문

## loop 반복문
- 반복적인 작업 수행
- while: 특정 조건 만족하는 경우 수행, 반복 중단 확인
- for: 컬렉션 타입의 아이템 순회하며 사용

```python
a = [1, 10, 9, 24, 566, 23, 45, 67 , 89]
```

```python
i = 0
while i < len(a):
    print(a[i])
    i+=1
```

```python
for item in a:
    print(item)
```

```python
for index, item in enumerate(a):
    print(index, item)
```

```python
for item in a:
    if item % 2: # 홀수인 경우
        print(item)
```

```python
while True:
    print("loop!")
```

## break, continue
- break: loop 중단
- continue: 아래 코드 실행하지 않고 조건으로 점프

```python
i = 0
while i < len(a):
    print(a[i])
    if a[i] > 20:
        break
    i+=1
```

```python
i = 0
while i < len(a):
    if a[i] > 40:
        continue
    print(a[i])
    i+=1
```

```python
for ch in "Hello world":
    print(ch)
```

```python
for x in 10: # 순회 가능한 항목만 가능
    print(x)
```

## dict 아이템 출력하기

```python
a = {"Korea": "Seoul", "Japan": "Tokyo", "China": "Beijing"}
```

```python
for key in a:
    print(key)
```

```python
for val in a.values():
    print(val)
```

```python
for key, val in a.items():
    print(key, val)
```

## for에서 index 사용

```python
b = [1, 2, 3, 4, 5]
for i, val in enumerate(b):
    print(i, val)
```

```python
a = {"Korea": "Seoul", "Japan": "Tokyo", "China": "Beijing"}
for i, (key, val) in enumerate(a.items()):
    print(i, key, val)
```

## 중첩 반복문

```python
for i in range(2, 10):
    for j in range(1, 10):
        print(i, ' * ', j, ' = ', i * j)
```

## 연습

```python
list(range(5, 101, 5))
```

### 1 ~ 100 2의 배수 또는 11의 배수 출력

```python
for i in range(1, 100):
    if i % 2 == 0 or i % 11 == 0:
        print(i)
```

### 최대, 최소 찾기

```python
a = [22, 1, 3, 4, 7, 98, 21, 55, 87, 99, 19, 20, 45]
min = a[0]
max = a[0]
for v in a:
    if v < min:
        min = v
    if v > max:
        max = v
        
print("min", min)
print("max", max)
del min, max
```

```python
a = [22, 1, 3, 4, 7, 98, 21, 55, 87, 99, 19, 20, 45]
print(min(a), max(a))
```

### 평균 구하기

```python
a = [22, 1, 3, 4, 7, 98, 21, 55, 87, 99, 19, 20, 45]
sum = 0
for v in a:
    sum += v
print("AVG", sum/len(a))
```
